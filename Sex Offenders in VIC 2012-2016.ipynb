```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dt=pd.read_csv('sexoffenders.csv')
```

```python
dt.shape
```

```python
dt
```

```python
dt['AgeGroup']=dt.AgeGroup.replace('Oct-14', '10-14')
dt['AgeGroup']=dt.AgeGroup.replace('65 years and over', '65+')
dt.columns = ['gender', 'AgeGroup', '2012', '2013', '2014', '2015', '2016']
```

```python
dt['Mean']=[(dt[['2012', '2013', '2014', '2015', '2016']].iloc[i].mean()) for i in range(dt.shape[0])]
```

```python
dt
```

```python
sortedage=dt.sort_values('AgeGroup')
sortedage
```

```python
plt.figure(figsize=(9,6))
plt.title('Sexual Offenders (from 2012 to 2016)')
sns.barplot(x = 'AgeGroup', y = "Mean", data = sortedage, hue='gender', )

plt.show()
```
